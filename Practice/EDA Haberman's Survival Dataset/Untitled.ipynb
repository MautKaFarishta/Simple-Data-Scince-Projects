```python
sns.set_style("whitegrid");
sns.pairplot(df, hue="status", height=3);
plt.show()
```

```python
#From above pair-plot we come to know that age is more likely to be most important feature


sns.set_style("whitegrid");
sns.FacetGrid(df, hue="status", height=3) \
   .map(plt.scatter, "nodes", "age") \
   .add_legend();
plt.show();

#We are trying to find out the relation between the nodes ,age and the status of the patient 
```

```python
df[df['age'] < 40].groupby('status').size()
```

# Observations

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='status',height=5)\
   .map(sns.distplot,'nodes')\
   .add_legend()
plt.title("Node PDF",size=20)
plt.show()

#Plotting PDF for nodes
#Less the number of node high chances of surviving
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='status',height=5)\
   .map(sns.distplot,'age')\
   .add_legend()
plt.title("Age PDF",size=20)
plt.show()

#Age PDF tells us that maximum patients are from range 40-60 years
```

```python
import numpy as np
counts, bin_edges = np.histogram(df['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.title("CDF ",size=20)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(df['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.title("CDF ",size=20)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

plt.xlabel("Nodes   Age")
# plt.legend()
```
